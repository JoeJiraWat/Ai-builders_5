```python
import os
from pydub import AudioSegment

def convert_ogg_to_wav_in_folder_limited(input_folder, output_folder, limit=2000):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
        print(f"สร้างโฟลเดอร์ '{output_folder}' สำหรับไฟล์ .wav เรียบร้อยแล้ว")

    print(f"กำลังสแกนหาไฟล์ .ogg ในโฟลเดอร์: {input_folder} (จะแปลงสูงสุด {limit} ไฟล์)")

    ogg_files_found = []
    for filename in os.listdir(input_folder):
        if filename.lower().endswith(".ogg"):
            ogg_files_found.append(filename)
    
    # เลือกเฉพาะ 2000 ไฟล์แรกที่พบ (หรือน้อยกว่าถ้ามีไม่ถึง)
    files_to_convert = ogg_files_found[:limit]
    
    converted_count = 0
    for filename in files_to_convert:
        ogg_filepath = os.path.join(input_folder, filename)
        wav_filename = os.path.splitext(filename)[0] + ".wav"
        wav_filepath = os.path.join(output_folder, wav_filename)

        try:
            print(f"กำลังแปลงไฟล์ ({converted_count + 1}/{len(files_to_convert)}): {filename}...")
            audio = AudioSegment.from_ogg(ogg_filepath)
            audio.export(wav_filepath, format="wav")
            print(f"แปลงสำเร็จ: {filename} -> {wav_filename}")
            converted_count += 1
        except Exception as e:
            print(f"เกิดข้อผิดพลาดในการแปลงไฟล์ {filename}: {e}")

    print(f"\nการแปลงไฟล์ .ogg ทั้งหมด {converted_count} ไฟล์เสร็จสมบูรณ์!")
    if len(ogg_files_found) > limit:
        print(f"มีไฟล์ .ogg ที่ข้ามไป {len(ogg_files_found) - limit} ไฟล์ เนื่องจากเกินจำนวนที่กำหนดไว้ ({limit} ไฟล์)")

if __name__ == "__main__":
    input_directory = "/kaggle/input/sakura-uta-voice/sakura-uta-voice/"  # เปลี่ยนเป็นชื่อโฟลเดอร์ที่มีไฟล์ .ogg ของคุณ
    output_directory = "/kaggle/working/sakura_2wav/" # เปลี่ยนเป็นชื่อโฟลเดอร์ที่คุณต้องการบันทึกไฟล์ .wav

    if not os.path.exists(input_directory):
        os.makedirs(input_directory)
        print(f"สร้างโฟลเดอร์ตัวอย่าง '{input_directory}' กรุณานำไฟล์ .ogg มาใส่ในโฟลเดอร์นี้")

    convert_ogg_to_wav_in_folder_limited(input_directory, output_directory, limit=2000)
```
